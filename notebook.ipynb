```python
from signal_processing_for_participant import load_empatica_signals, save_empatica_signals_to_csv

import matplotlib.pyplot as plt
```

```python
signals_b = load_empatica_signals(participant_id=3, condition='baseline')
print(signals_b['bvp'].shape, signals_b['eda'].shape, signals_b['temp'].shape)

# To stack into a design matrix s(t) âˆˆ R^3:
import numpy as np
s_t = np.vstack([signals_b["bvp"], signals_b["eda"], signals_b["temp"]]).T
print(s_t.shape)  # (time steps, 3)
```

```python
plt.figure(figsize=(15, 5))
plt.plot(s_t[:, 0], label='BVP')
plt.plot(s_t[:, 1], label='EDA')    
plt.plot(s_t[:, 2], label='TEMP')
plt.legend()
plt.title('Signals for Participant 3 - Baseline')
plt.xlabel('Time Steps')
plt.ylabel('Signal Value')
plt.show()
```
